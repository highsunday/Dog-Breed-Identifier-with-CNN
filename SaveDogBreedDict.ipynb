```python
import os
import numpy as np
from tqdm import tqdm
from glob import glob
from PIL import Image
import cv2                
import matplotlib.pyplot as plt                        
%matplotlib inline
```

```python
train_path='data/images/train'
```

```python
list_1=os.listdir(train_path)
DogBreedDir={}
count=0
for i in list_1:
    print(i.split("-")[1])
    DogBreedDir[i.split("-")[1]]=count
    count+=1
```

```python
DogBreedDir
```

```python
import csv

#儲存字典
with open('DogBreedDir.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['狗種', '編號'])
    for i in DogBreedDir.keys():
        #print(i,DogBreedDir[i])
        writer.writerow([i,DogBreedDir[i]])
```
