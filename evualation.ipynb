```python
from keras.models import load_model
import os
import numpy as np
from tqdm import tqdm
from glob import glob
from PIL import Image
import cv2                
import matplotlib.pyplot as plt                        
%matplotlib inline
```

```python
model = load_model('model/model_V7_50epochs.h5') 
```

```python
model.summary()
```

```python
test_path='data/images/test'
```

```python
list_1=os.listdir(test_path)
```

```python
DogBreedDir={}
count=0
for i in list_1:
    print(i.split("-")[1])
    DogBreedDir[i.split("-")[1]]=count
    count+=1
```

```python
All_Picture={}
img_size = 150  #根据实际情况可修改

for file in list_1:
    nub_train = len(glob(test_path + '/'+file+'/*.jpg'))
    X_train = np.zeros((nub_train,img_size,img_size,3),dtype=np.uint8)
    i = 0
    print(file)
    for img_path in tqdm(glob(test_path + '/'+file+'/*.jpg')):
        img = Image.open(img_path)
        img = img.resize((img_size,img_size)) #图片resize
        arr = np.asarray(img)  #图片转array
        X_train[i, :, :, :] = arr #赋值
        i += 1
    All_Picture[file.split('-')[1]]=X_train
```

```python
All_Picture
```

```python
AllData=[]
AllData_Label=[]
```

```python
AllBreed=All_Picture.keys()
#DogBreedDir :種類與數字的對應
```

```python
for i in AllBreed:
    print(i,':',DogBreedDir[i])
    temp=All_Picture[i]
    for j in temp:
        AllData.append(j)
        AllData_Label.append(DogBreedDir[i])
```

```python
AllData=np.array(AllData)
AllData_Label=np.array(AllData_Label)
OrignialData=AllData
```

```python
AllData.shape,AllData_Label.shape
```

```python
AllData=AllData/255
```

```python
AllData
```

```python
X=AllData
from keras.utils import np_utils
y = np_utils.to_categorical(AllData_Label, num_classes = 120)
```

```python
res=model.predict(X)
```

```python
#model.evaluate(x=AllData, y=AllData_Label, batch_size=64, verbose=1)
```

```python
res.shape
```

```python
#for i in range(len(res)):
count=0
for i in range(len(res)):
    print(res[i].argmax(),':',AllData_Label[i],':',res[i][AllData_Label[i]])
    if(res[i].argmax()==AllData_Label[i]):
        count=count+1
```

```python
print("準確率:",count/len(res))
```

```python
loss, accuracy = model.evaluate(X, y)
```

```python
DogBreedDir_reverse = dict(zip(DogBreedDir.values(), DogBreedDir.keys()) )
```

```python
DogBreedDir_reverse
```

```python
def returnDogBreed(num,inputImage):
    plt.imshow(inputImage)
    inputImage=inputImage/255
    inputImage=inputImage.reshape(1,150,150,3)
    res=model.predict(inputImage)
    #print(res[0])
    list1=[]
    for i in range(len(res[0])):
        list1.append([res[0][i],i])
    #print(list1)
    list1=sorted(list1, key=lambda x: x[0], reverse=True)
    #print(list1)
    for i in range(num):
        print(DogBreedDir_reverse[list1[i][1]],':',list1[i][0])
    return DogBreedDir_reverse[list1[i][1]]
```

```python
returnDogBreed(10,OrignialData[119])
print('正解為：',DogBreedDir_reverse[AllData_Label[119]])
```

```python
for i in range(10):
    if(returnDogBreed(0,OrignialData[i])==DogBreedDir_reverse[AllData_Label[i]]):
        print(DogBreedDir_reverse[AllData_Label[i]])
```
