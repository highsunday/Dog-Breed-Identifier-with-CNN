```python
from keras.models import load_model
import os
import numpy as np
from tqdm import tqdm
from glob import glob
from PIL import Image
import cv2                
import matplotlib.pyplot as plt                        
%matplotlib inline
```

```python
model=load_model('model_Test.h5')
model.summary()
```

```python
train_path='data/debug/test'
```

```python
list_1=os.listdir(train_path)
```

```python
DogBreedDir={}
count=0
for i in list_1:
    print(i.split("-")[1])
    DogBreedDir[i.split("-")[1]]=count
    count+=1
```

```python
All_Picture={}
img_size = 299  #根据实际情况可修改

for file in list_1:
    nub_train = len(glob(train_path + '/'+file+'/*.jpg'))
    X_train = np.zeros((nub_train,img_size,img_size,3),dtype=np.uint8)
    i = 0
    print(file)
    for img_path in tqdm(glob(train_path + '/'+file+'/*.jpg')):
        img = Image.open(img_path)
        img = img.resize((img_size,img_size)) #图片resize
        arr = np.asarray(img)  #图片转array
        X_train[i, :, :, :] = arr #赋值
        i += 1
    All_Picture[file.split('-')[1]]=X_train
```

```python
plt.imshow(All_Picture['Sealyham_terrier'][0])
```

```python
AllData=[]
AllData_Label_list=[]
```

```python
AllBreed=All_Picture.keys()
#DogBreedDir :種類與數字的對應
```

```python
for i in AllBreed:
    print(i,':',DogBreedDir[i])
    temp=All_Picture[i]
    for j in temp:
        AllData.append(j)
        AllData_Label_list.append(DogBreedDir[i])
```

```python
AllData=np.array(AllData)
AllData_Label=np.array(AllData_Label_list)
```

```python
AllData.shape,AllData_Label.shape
```

```python
from keras.utils import np_utils
AllData_Label = np_utils.to_categorical(AllData_Label, num_classes = 2)
```

```python
AllData_Label
```

```python
AllData
```

```python
X_test=[]
y_test=[]
```

```python
last=np.zeros(2)

for i in range(len(AllData)):
    if(not(last==AllData_Label[i]).all()):
        last=AllData_Label[i] 
    X_test.append(AllData[i])
    y_test.append(AllData_Label[i])
```

```python
X_test=np.array(X_test)
y_test=np.array(y_test)
```

```python
X_test=X_test/255
```

```python
X_test
```

```python
len(X_test[0])
```

```python
preds = model.predict(X_test)
```

```python
preds
```

```python
for i in range(len(preds)):
    if(preds[i][0]>preds[i][1]):
        if(AllData_Label_list[i]==1):
            print(i,' 分錯類')
    else:
        if(AllData_Label_list[i]==0):
            print(i,' 分錯類')
```
